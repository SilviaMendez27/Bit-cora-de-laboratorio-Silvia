### Operación con lista de datos

### Promedios

```python
datos = [10, 12, 15, 18, 20]
```

```python
promedio = sum(datos)/len(datos)
```

```python
print(promedio)
```

```python
datos_dos = [5, 8, 12, 9, 10, 15, 18, 20, 21, 3]
```

```python
promedio_dos = sum(datos_dos)/len(datos_dos)
```

```python
print(promedio_dos)
```

### Numero maximo

```python
maximo = max(datos_dos)
```

```python
print(maximo)
```

### Promedio

```python
def calcular_promedio(lista):
    return sum(lista)/len(lista)
```

```python
lista=[4,8,3,7,9]
```

```python
print (calcular_promedio(lista))
```

### Valor maximo

```python
def valor_maximo(lista):
    return max(lista)
```

```python
print(valor_maximo(lista))
```

### Valor minimo

```python
def valor_minimo(lista):
    return min(lista)
```

```python
print (valor_minimo(lista))
```

```python
lista_dos=[12, 15, 18, 22, 45]
```

```python
def valor_min(lista_dos): 
    return min(lista_dos)
```

```python
print(valor_min(lista_dos))
```
