### Importar libreria

```python
import matplotlib.pyplot as plt
```

### Gráfica de líneas

```python
datos = [3, 5, 7, 4, 6]
```

```python
plt.plot(datos)
plt.title("Datos simulados")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.show()
```

```python
datos2 = [27,24,17,32,12,17,23,19,28,11]
```

```python
plt.plot(datos2)
plt.title("Nueva gráfica")
plt.show()
```

### Gráfica de barras

```python
carreras = ["Medicina", "Enfermería", "Odontología"]
estudiantes = [50, 35, 20]

plt.bar(carreras, estudiantes)
plt.title("Número de estudiantes por carrera")
plt.xlabel("Carrera")
plt.ylabel("Cantidad")

plt.show()
```

### Gráfica de pastel

```python
frutas = ["Manzana", "Plátano", "Naranja"]
cantidad = [40, 35, 25]

plt.pie(cantidad, labels=frutas, autopct="%1.1f%%")
plt.title("Distribución de frutas")

plt.show()
```

### Histograma

```python
calificaciones = [60,70,80,90,75,85,95,65,70,88]

plt.hist(calificaciones)
plt.title("Distribución de calificaciones")
plt.xlabel("Calificación")
plt.ylabel("Frecuencia")

plt.show()
```

### Gráfica de área

```python
x = [1,2,3,4,5]
y = [3,5,4,6,7]

plt.fill_between(x, y)
plt.title("Gráfica de área")
plt.xlabel("Tiempo")
plt.ylabel("Valor")

plt.show()
```

### Gráfica de bigotes

```python
datos = [10,12,14,15,18,20,22,25,30]

plt.boxplot(datos)
plt.title("Distribución de datos")

plt.show()
```

### Gráfica 3D

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Crear datos
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)

X, Y = np.meshgrid(x, y)

Z = np.sin(np.sqrt(X**2 + Y**2))

# Crear gráfica
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z)

ax.set_title("Superficie 3D")

plt.show()
```
