### Gráfica 3D

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Crear datos
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)

X, Y = np.meshgrid(x, y)

Z = np.sin(np.sqrt(X**2 + Y**2))

# Crear gráfica
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z)

ax.set_title("Superficie 3D")

plt.show()
```

# Introducción al procesamiento digital de imagenes 

Importar libreria

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
```

## Cargar imagen lena

```python
# Cargar imagen
url='https://itc.ua/wp-content/uploads/2024/04/Lenna_test_image.png'
resp=urllib.request.urlopen(url)
image=np.asarray(bytearray(resp.read()),dtype="uint8")
imagen=cv2.imdecode(image,cv2.IMREAD_GRAYSCALE)

# Mostrar imagen


plt.imshow(imagen, cmap='gray')
plt.title('Lenna')
plt.axis("off")
plt.show()
```

## Lena horizontal

```python
A = imagen
B = np.zeros((512,512)) # Matriz 512x512 de 0
for i in range(512):
    for j in range(512):
        B[512 - 1 - j, i] = A[i, j]; # Lena Horizontal

plt.imshow(B, cmap = 'gray')
plt.title('Lena Horizontal')
plt.axis("off")
plt.show()
```

## Tamaño de imagen

```python
imagen.shape
```

## Lena de cabeza

```python
A = imagen
C = np.zeros((512,512)) # Matriz 512x512 de 0
for i in range(512):
    for j in range(512):
        C[512 - 1 - i, j] = A[i, j]; # Lena de Cabeza
import matplotlib.pyplot as plt
plt.imshow(C, cmap = 'gray')
plt.title('Lena de Cabeza')
plt.axis("off")
plt.show()
```

## Recorte de lena

```python
#Recorte una porción de su imagen que muestre principalmentesu cabeza. Imprima el resultado y los parámetros utilizados.

N1, N2 = 64, 128 # Tamaño N1=64 y N2=128 (dimensiones del recorte en N)
E = np.zeros((64,128)) # Matriz de 64x128 en 0
for k in range(N1): # k va de 0 a 63
    for l in range(N2): # l va de 0 a 127
        E[k, l] = A[255 + k, 255 + l] # 255 define el punto de inicio desde el␣,→cual se extraen los valores para llenar E
plt.imshow(E, cmap='gray')
plt.title('Recorte Lenna')
plt.axis("off")
plt.show()
```

## Media

```python
suma_total = 0 # acumulará la suma de todos los elementos de la matriz A
num_elementos = 0 # contará el número total de elementos en la matriz A
m, n = A.shape # m filas, n columnas

for i in range(m):
    for j in range(n):
        suma_total += int(A[i,j]) # se añade el valor de cada elemento de la matriz A
        num_elementos += 1 # se incrementa 1 al contador por cada elemento␣,→visitado en la matriz A
media = suma_total / num_elementos
print("Media:", media)
```
