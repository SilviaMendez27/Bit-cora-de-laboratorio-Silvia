## Parte 2 - Función de clasificación clínica 

**Clasificación según su nivel de glucosa**

```python
def clasificar_glucosa(valor):
    if valor<100:
        return "Normal"
    elif 100<=valor<=125:
        return "Prediabetes"
    else: 
        return "Diabetes"
    
```

**Clasificación según su nivel de hba1c**

```python
def clasificar_hba1c(valor):
    if valor<5.7:
        return "Normal"
    elif 5.7<=valor<=6.4:
        return "Prediabetes"
    else: 
        return "Diabetes"
```

**Clasificación de una lista de niveles de glucosa**

```python
def clasificar_lista(lista):
    conteo={"Normal":0, "Prediabetes":0, "Diabetes":0}
    clasificaciones=[]
    
    for valor in lista: 
        categoria=clasificar_glucosa(valor)
        clasificaciones.append(categoria)

        if categoria=="Normal":
            conteo["Normal"]+=1
        elif categoria=="Prediabetes":
            conteo["Prediabetes"]+=1
        else:
            conteo["Diabetes"]+=1
            
    return clasificaciones, conteo
```

```python
print(clasificar_lista(glucosa))
```

## Parte 3 - Función para graficar 

**Grafica sin librería**

```python
def grafica_histograma(datos, titulo, xlabel):
    print(titulo)
    print(xlabel)

    minimo=min(datos)
    maximo=max(datos)

    bins=5
    rango=(maximo-minimo)/ bins

    frecuencias=[0]*bins

    #Contar datos en intervalos
    for valor in datos:
        indice=int((valor-minimo)/rango)
        if indice ==bins:#caso borde
            indice-=1
        frecuencias[indice]+=1

        #Mostrar histograma 
        for i in range(bins):
            inicio=minimo+i*rango
            fin=inicio+rango
            print(f"{inicio:.1f}-{fin:.1f} | {'*' * frecuencias[i]}")
    
    
```

```python
print (grafica_histograma(glucosa, "Histograma de Glucosa", "mg/dL"))
```

```python
print (grafica_histograma(hba1c, "Histograma de hba1c", "mg/dL"))
```

## Parte 4 - Correlación simple

**Calculo de r e interpretación de correlación**

```python
def correlación_simple(x, y):
    n=len(x)

    #Medias
    media_x=sum(x)/n
    media_y=sum(y)/n

    #Sumatorias
    numerador=0
    suma_x=0
    suma_y=0

    for i in range(n):
        dx=x[i]-media_x
        dy=y[i]-media_y

        numerador += dx*dy
        suma_x+=dx**2
        suma_y+=dy**2

    denominador=(suma_x*suma_y)**0.5

    r=numerador/denominador

    #Interpretacion
    if abs(r)>=0.7:
        interpretacion="Fuerte"
    elif abs(r)>=0.4:
        interpretación = "Moderna"
    else:
        interpretación ="Debil"

    return r, interpretacion
```

```python
print (correlación_simple(glucosa, hba1c))
```
