# Tarea: Procesamiento de Imagen Médica

## Instrucciones

Importar libreria

```python
import cv2
import matplotlib.pyplot as plt
```

**2. Leer una imagen**

```python
img = cv2.imread("uva.jpg")
```

**3. Mostrar la imagen original**

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

**4. Convertir la imagen a escala de grises**

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray, cmap='gray')
plt.title("Imagen en escala de grises")
plt.show()
```

**5. Comparar la imagen original con la imagen en escala de grises**

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Imagen original")

plt.subplot(1,2,2)
plt.imshow(gray, cmap='gray')
plt.title("Escala de grises")

plt.show()
```

**6. Calcular y visualizar el histograma de la imagen en escala de grises**

```python
plt.hist(gray.ravel(), bins=256)
plt.title("Histograma")
plt.show()
```

**7. Interpretar qué información proporciona el histograma**

En el histograma se muestra la ditribución de intensidades color en la imagen, se observa un pico principal aproximadamente en 150, interpretandolo como el grupo principal de intensidad. La mayor concentración de pixeles está entre 50 y 170, por lo tanto, se determina que la imagen contiene principalmente grises medios. Al estar los extremos con valores no tan altos se concluye que no hay demasiados píxeles completamente negros ni blancos.

**8. Aplicar una técnica de umbralización para segmentar la imagen**

```python
_, thresh = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh, cmap='gray')
plt.title("Imagen umbralizada")
plt.show()
```

```python
_, thresh = cv2.threshold(gray, 50, 170, cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh, cmap='gray')
plt.title("Imagen umbralizada dos")
plt.show()
```

**9. Guardar la imagen resultante de la umbralización**

```python
cv2.imwrite("resultado_umbral_uva.jpg", thresh)
```

## Preguntas de análisis

**1 ¿Qué diferencias observas entre la imagen original y la imagen en escala de grises?**

La diferencia radica en los colores, la imagen en escala de grises no tiene colores, solo muestra tonos de gris, desde negro hasta blanco, dado que contiene un solo canal de intensidad. De manera contraria, la imagen original tiene colores, dado que usa los canales rojo, verde y azul, lo que le permite representar una amplia gama de colores.

**2 ¿Qué representa el histograma de una imagen?**

El histograma representa la distribución de las intensidades de los píxeles dentro de la imagen. En el eje x se colocan los niveles de intensidad en una escala de 0 a 255, donde 0 es negro y 255 blanco. En el eje y se colocan los números de píxeles que poseen esa intensidad. En el histograma es posible analizar características de la imagen como el nivel de brillo, contraste y el predominio de zonas claras u oscuras. 

**3 ¿Qué efecto tiene la umbralización sobre la imagen?**

La umbralización se encarga de transformar una imagen en escala de grises en una a blanco y negro (imagen binaria), esto se logra estableciendo un valor de umbrar. Los píxeles con intensidad mayor al umbral se convierten en blancos y los píxeles con intensidad menor al umbral se convierten en negros, de tal manera que ayuda a separar objetos del fondo. 

**4 ¿Cómo podría utilizarse esta técnica en el análisis de imágenes médicas?**

Esta técnica es utilizada para identificar estructuras específicas en estudios. Es posible utilizarla para segmentar tumores o lesiones, identificar tejidos específicos, delimitar órganos, detectar anomalías o zonas inflamadas, etc. De tal manera que facilita el análisis.
