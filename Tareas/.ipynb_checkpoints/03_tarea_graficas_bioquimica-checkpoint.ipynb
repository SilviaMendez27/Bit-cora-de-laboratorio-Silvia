# Tarea: Visualización de Datos Bioquímicos con Matplotlib

### Objetivo:
-Visualizar datos bioquímicos mediante herramientas gráficas para identificar patrones, tendencias y relaciones que faciliten su análisis e interpretación

## Parte A - Figuras obligatorias

**Importar libreria**

```python
import matplotlib.pyplot as plt
```

### 1) Histograma de glucosa 

```python
#Datos
glucosa = [82, 95, 110, 130, 88, 76, 145, 99, 102, 118, 90, 104, 127, 140, 85]

#Histograma, glucosa=datos, bins=intervalos
plt.hist(glucosa, bins=8)

#Titulo 
plt.title("Distribución de niveles de glucosa")

#Ejes
plt.xlabel("Glucosa (mg/dL)")
plt.ylabel("Frecuencia")

#Grid: agregar cuadricula
plt. grid(True)

#Muestra la figura
plt.show()
```

Los bins representan los intervalos en los que se agrupan los datos dentro de un histograma para visualizar su distribución.

El rango que se encuentra más frecuente es de 75 a 92 mg/dL y de 102 a 110 mg/dL, se observan valores altos de 80, 100 y 140 mg/dL

### 2) Gráfica de barras (conteo por grupo)

```python
categorias = ["Normal", "Prediabetes", "Diabetes"]
valores = [7, 4, 4]

plt.bar(categorias, valores)
plt.title("Clasificaciones de niveles de glucosa")
plt.xlabel("Ctegoria")
plt.ylabel("Cantidad")

#numeros arriba de las barras
for i in range(len(valores)):
    plt.text(i, valores[i] + 0.1, valores[i], ha='center')


plt.show()
```

En la grafica se puede observar que el grupo dominante es el "normal" con un valor de 7 datos 

### 3) Scatter: glucosa vs HbA1c

```python
hba1c = [5.1, 5.4, 5.9, 6.7, 5.2, 5.0, 7.2, 5.5, 5.8, 6.1, 5.3, 5.9, 6.6, 7.0, 5.1]

#Crear lista para colores
colores = []

#Clasificar cada valor para glucosa
for valor in glucosa:
    #Condiciones
    if valor < 100:
        colores.append("pink")      # Normal
    elif valor < 126:
        colores.append("yellow")     # Prediabetes
    else:
        colores.append("red")        # Diabetes

#Crear grafica de dispersion
plt.scatter(glucosa, hba1c, c=colores)

plt.title("Relación entre glucosa y HbA1c")
plt.xlabel("Glucosa (mg/dL)")
plt.ylabel("HbA1c (%)")

plt.grid(True)

#Crear leyenda
import matplotlib.patches as mpatches
#Definir categorias
normal = mpatches.Patch(color='pink', label='Normal')
pred = mpatches.Patch(color='yellow', label='Prediabetes')
diab = mpatches.Patch(color='red', label='Diabetes')

#Mostrar leyenda
plt.legend(handles=[normal, pred, diab])

plt.show()
```

Es posible observar que conforme aumenta el valor de la glucosa también aumenta el valor de hbA1c, indicando una relación positiva entre las variables

### 4)Figura mejorada (versión publicable)

```python
colores = []

for valor in glucosa:
    if valor < 100:
        colores.append("pink")
    elif valor < 126:
        colores.append("yellow")
    else:
        colores.append("red")

#Crear la figura
plt.figure(figsize=(8,6))

plt.scatter(glucosa, hba1c, c=colores)

plt.title("Relación entre niveles de glucosa en sangre y hemoglobina glucosilada (HbA1c)")
plt.xlabel("Glucosa en sangre (mg/dL)")
plt.ylabel("HbA1c (%)")

#Ajustar limite de los ejes
plt.xlim(70,150)
plt.ylim(4.8,7.5)

plt.grid(True)

normal = mpatches.Patch(color='pink', label='Normal')
pred = mpatches.Patch(color='yellow', label='Prediabetes')
diab = mpatches.Patch(color='red', label='Diabetes')

plt.legend(handles=[normal, pred, diab])

#Guardar imagen, dpi alta resolución
plt.savefig("glucosa_hba1c.png", dpi=300)

plt.show()
```

## Parte B - Reto

### 5) Agregar lineas de referencia clinica

```python
colores = []

for valor in glucosa:
    if valor < 100:
        colores.append("pink")
    elif valor < 126:
        colores.append("yellow")
    else:
        colores.append("red")

plt.figure(figsize=(8,6))

plt.scatter(glucosa, hba1c, c=colores)

plt.title("Relación entre niveles de glucosa en sangre y hemoglobina glucosilada (HbA1c)")
plt.xlabel("Glucosa en sangre (mg/dL)")
plt.ylabel("HbA1c (%)")

plt.xlim(70,150)
plt.ylim(4.8,7.5)

plt.grid(True)

normal = mpatches.Patch(color='pink', label='Normal')
pred = mpatches.Patch(color='yellow', label='Prediabetes')
diab = mpatches.Patch(color='red', label='Diabetes')

plt.legend(handles=[normal, pred, diab])

plt.savefig("glucosa_hba1c.png", dpi=300)


# Líneas de referencia glucosa
plt.axvline(x=100, color='orange', linestyle='--')
plt.axvline(x=126, color='brown', linestyle='--')

# Líneas de referencia HbA1c
plt.axhline(y=5.7, color='orange', linestyle='--')
plt.axhline(y=6.5, color='brown', linestyle='--')


plt.show()
```

Se observa en los datos que 8 valores de glucosa superan los 100 mg/dL y 4 superan los 126 mg/dL. En el caso de HbA1c, 8 valores superan el 5.7 % y 4 superan el 6.5 %
