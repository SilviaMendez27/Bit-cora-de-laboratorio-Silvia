Parte 3 - Promedio y operaciones con listas

```python
Calificaciones=[85,90,78,92,88]
```

```python
promedio=sum(Calificaciones)/len(Calificaciones)
```

```python
print(promedio)
```

```python
maximo=max(Calificaciones)
```

```python
print(maximo)
```

```python
minimo=min(Calificaciones)
```

```python
print(minimo)
```

```python
diferencia=max(Calificaciones)-min(Calificaciones)
```

```python
print(diferencia)
```

Parte 4 - Introducción a vectores 

```python
v1=[1,2,3]
v2=[4,5,6]
```

```python
suma=[v1[i]+v2[i] for i in range(len(v1))]
```

```python
print(suma)
```

```python
resta = [v1[i] - v2[i] for i in range(len(v1))]
```

```python
print(resta)
```

```python
producto = [v1[i] * v2[i] for i in range(len(v1))]
```

```python
print(producto)
```
