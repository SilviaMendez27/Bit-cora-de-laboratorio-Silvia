# Filtrado digital: Canny Edge Detection 

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import os
import cv2
```

### Peorcitas

```python
# Ruta de la carpeta de imágenes
carpeta_imagenes = '/Users/silviamendez/Desktop/Servicio Social/Tareas pdf/Imagenes'


# Filtrar solo archivos de imagen (puedes ajustar según el formato que uses: .jpg, .png, etc.)
formatos_validos = (".jpg", ".png", ".jpeg")
```

```python
# Directorio de entrada y salida
input_dir = '/Users/silviamendez/Desktop/Servicio Social/Tareas pdf/Imagenes'  # Cambia esta ruta a la ubicación de tu carpeta de imágenes
output_dir = '/Users/silviamendez/Desktop/Servicio Social/Salida canales'   # Cambia esta ruta a donde quieras guardar las imágenes procesadas
```

```python
# Parámetros de Canny Edge Detection, controla que tan sensible serán los bordes detectados
lower_threshold = 30 #Detecta más bordes
upper_threshold = 70 #Detecta menos bordes
```

```python
# Fuente para los títulos
font = cv2.FONT_HERSHEY_SIMPLEX #Tipo de letra
font_scale = 0.6 #Tamaño
font_color = (255,)  # Blanco en escala de grises
thickness = 1 #Grosor

# Procesar cada imagen
for filename in os.listdir(input_dir): #Recorre todas las imágenes
    if filename.endswith('.jpg') or filename.endswith('.png'): #Filta imagenes
        img_path = os.path.join(input_dir, filename) #Une la ruta y el nombre
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE) #Carga la imagen en escala de grises

       #Verificar que se cargó correctamente 
        if img is None:
            print(f"Error al cargar la imagen: {filename}")
            continue

        # Suavizado, reducción de ruido 
        blurred_img = cv2.GaussianBlur(img, (5, 5), 1.5)

        # Bordes Canny
        edges = cv2.Canny(blurred_img, lower_threshold, upper_threshold)

        # Crear copias para poner títulos y no modificar la original 
        img_with_title = img.copy()
        edges_with_title = edges.copy()

        # Agregar títulos
        cv2.putText(img_with_title, 'Original', (10, 25), font, font_scale, font_color, thickness, cv2.LINE_AA)
        cv2.putText(edges_with_title, 'Canny', (10, 25), font, font_scale, font_color, thickness, cv2.LINE_AA)

        # Combinar lado a lado
        combined = np.hstack((img_with_title, edges_with_title))

        # Guardar
        output_path = os.path.join(output_dir, f"original_canny_{filename}")
        cv2.imwrite(output_path, combined)
        print(f"Imagen combinada con títulos guardada en: {output_path}")

print("Procesamiento completado.")
print("Procesamiento completado.")
```
