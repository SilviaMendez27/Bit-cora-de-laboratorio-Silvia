### Entrenamiento del algorítmo 

#### 1. Probar o rechazar la hipótesis de que la distribución de frecuencias de la columna BMI sigue una distribución normal.

 El test de Shapiro-Wilk es una prueba estadística que se utiliza para determinar si una muestra de datos se ajusta a una distribución normal. Es una de las pruebas más comunes y eficaces para este propósito, especialmente recomendada para muestras pequeñas (generalmente inferiores a 50 elementos, aunque puede usarse con muestras de hasta 2000).
 
 #### Hipótesis:

1. **Hipótesis nula (H0)**: La muestra proviene de una población con una distribución normal.
1. **Hipótesis alternativa (H1)**: La muestra no proviene de una población con una distribución normal.

```python
from scipy.stats import shapiro

# Realizar la prueba de Shapiro-Wilk para la normalidad
shapiro_test = shapiro(heart_data['BMI'].sample(5000))  # Muestreo debido a limitaciones de tamaño en el test

display(shapiro_test)
```

Dado que el p-valor es mucho menor que el nivel de significancia típico (por ejemplo, 0.05), hay evidencia suficiente para rechazar la hipótesis nula. Esto significa que los datos de BMI no siguen una distribución normal según la prueba de Shapiro-Wilk.

Por lo tanto, la prueba de Shapiro-Wilk para la columna BMI arroja un estadístico de prueba de aproximadamente 0.932 y un p-valor extremadamente pequeño (prácticamente cero). Este p-valor indica que podemos **rechazar la hipótesis nula de que los datos de BMI siguen una distribución normal.**


#### 2 . Probar o rechazar la hipótesis de que las variables BMI y GenHealth son independientes. Utilice la siguiente escala para realizar la tabla de contingencia y la columna HeartDisease con 1 = `Yes HeartDisease` y 0 = `No HeartDisease`.(25.3 Kg/m²,Overweight)

 #### Hipótesis:

1. **Hipótesis nula (H0)**:  BMI y GenHealth son independientes.
1. **Hipótesis alternativa (H1)**:  BMI y GenHealth no son independientes.

```python
from scipy.stats import chi2_contingency

# Crear una nueva columna para categorizar BMI
heart_data['BMI_Category'] = heart_data['BMI'].apply(lambda x: '>=25.3' if x >= 25.3 else '<25.3')

# Crear la tabla de contingencia
contingency_table = pd.crosstab(heart_data['BMI_Category'], heart_data['GenHealth'])

# Realizar el test de Chi-cuadrado
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(chi2, p, dof)
display(contingency_table)
```

El test de Chi-cuadrado para la tabla de contingencia entre las categorías de BMI y los niveles de GenHealth resulta en un estadístico de Chi-cuadrado de aproximadamente 1562.03, con un p-valor de 0.0. Esto indica que hay suficiente evidencia para **rechazar la hipótesis nula de que BMI y GenHealth son independientes**, sugiriendo que estas dos variables están asociadas.

#### 3. Obtenga la matriz de confusión para cada uno de los métodos de regresión logística, svm y decision tree
¿Cuál de los 3 algoritmos obtiene la mejor exactitud?
¿Cuál de los 3 algoritmos seleccionaría y por qué?


#### División de datos 

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

# Codificar variables categóricas
for column in heart_data.columns:
    if heart_data[column].dtype == 'object':
        heart_data[column] = label_encoder.fit_transform(heart_data[column])

X = heart_data.drop('HeartDisease', axis=1)
y = heart_data['HeartDisease']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

##### Entrenamiento de modelos

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

# Regresión logística
log_reg = LogisticRegression(random_state=42, max_iter=1000)
log_reg.fit(X_train, y_train)
log_pred = log_reg.predict(X_test)

# SVM
svm_model = SVC()
svm_model.fit(X_train, y_train)
svm_pred = svm_model.predict(X_test)

# Árbol de decisión
tree_model = DecisionTreeClassifier()
tree_model.fit(X_train, y_train)
tree_pred = tree_model.predict(X_test)
```

#### Matriz de confusión y exactitud para cada modelo


```python

# Regresión logística
log_conf = confusion_matrix(y_test, log_pred)
log_acc = accuracy_score(y_test, log_pred)

# SVM
svm_conf = confusion_matrix(y_test, svm_pred)
svm_acc = accuracy_score(y_test, svm_pred)

# Árbol de decisión
tree_conf = confusion_matrix(y_test, tree_pred)
tree_acc = accuracy_score(y_test, tree_pred)

```

##### Visualización de las matrices de confusión 

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import numpy as np

def plot_confusion_matrix(y_true, y_pred, title='Confusion Matrix'):
    # Calcular la matriz de confusión
    conf_matrix = confusion_matrix(y_true, y_pred)
    fig, ax = plt.subplots(figsize=(6, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=np.unique(y_true), yticklabels=np.unique(y_true))
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.title(title)
    plt.show()
```

```python
# De acuerdo a las predicciones log_pred, svm_pred, tree_pred
plot_confusion_matrix(y_test, log_pred, title='Logistic Regression Confusion Matrix')
print('Acurracy de la regresión logística:',log_acc)
plot_confusion_matrix(y_test, svm_pred, title='SVM Confusion Matrix')
print('Acurracy de la máquina de soporte vectorial:',svm_acc)
plot_confusion_matrix(y_test, tree_pred, title='Decision Tree Confusion Matrix')
print('Acurracy del árbol de decisiones:',tree_acc)
```

#### 1. ¿Cuál de los 3 algoritmos obtiene la mejor exactitud?

Maquina de soporte vectorial 

#### 1. ¿Cuál de los 3 algoritmos seleccionaría y por qué?


Primero que nada la Regresión Logística ha logrado clasificar correctamente la mayoría de los casos negativos pero ha tenido dificultades con los positivos, con muchos falsos negativos,SVM ha clasificado todos los casos como negativos. Aunque ha evitado los falsos positivos, ha fallado en identificar cualquier caso positivo, resultando en un alto número de falsos negativos,el Árbol de Decisiones ha mostrado un mejor equilibrio en la identificación de casos positivos comparado con los otros modelos, aunque a costa de un aumento en los falsos positivos.

Aunque el **SVM** tiene la mayor exactitud, su incapacidad para identificar cualquier caso positivo es preocupante, especialmente si la variable `HeartDisease` representa una condición seria donde los falsos negativos pueden tener consecuencias graves. El **Árbol de Decisiones**, aunque menos preciso en general, podría ser preferible si es importante minimizar los falsos negativos a costa de algunos falsos positivos.
Sin embargo, si se busca un balance general con una buena exactitud y una tasa aceptable de falsos negativos, la **Regresión Logística** podría ser la mejor opción.
